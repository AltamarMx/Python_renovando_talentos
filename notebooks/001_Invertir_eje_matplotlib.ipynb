```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = 4 + x*3

# plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)
ax.set_title('profundidad')

plt.show()
```

```python

# plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)
ax.set_ylim(35,5)
ax.set_title('profundidad')

plt.show()
```
